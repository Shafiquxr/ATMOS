```python
%%writefile valvar.l
%{
#include "y.tab.h"
%}
%%
[a-zA-Z] {return LET;}
[0-9] {return DIG;}
. {return yytext[0];} /* This rule catches any other single character */
\n {return 0;}
%%
int yywrap() {
return 1;
}
```

```python
%%writefile valvar.y
%{
#include<stdio.h>
%}
%token LET DIG
%%
variable: var ;
var: var DIG | var LET | LET ;
%%
int main() {
printf("Enter the variable:\n");

yyparse();
printf("Valid variable\n");

return 0;
}
int yyerror() {
printf("Invalid variable\n");

exit(0);
}
```

### LEX and YACC Code for Arithmetic Expression Parsing

```python
%%writefile art_expr.l
%{
#include<stdio.h>
#include "y.tab.h"
%}
%%
[a-zA-Z][0-9a-zA-Z]* {return ID;}
[0-9]+ {return DIG;}
[ \t]+ {;}
. {return yytext[0];}
\n {return 0;}
%%
int yywrap() {
return 1;
}
```

```python
%%writefile art_expr.y
%{
#include<stdio.h>
#include<stdlib.h> // Required for exit()
%}
%token ID DIG
%left '+' '-'
%left '*' '/'
%right UMINUS
%%
stmt: expn
;

expn: expn '+' expn | expn '-' expn | expn '*' expn | expn '/' expn | '-' expn %prec UMINUS | '(' expn ')' | DIG | ID;
%%
int main() {
printf("Enter the Expression\n");

yyparse();
printf("valid Expression\n");

return 0;
}
int yyerror() {
printf("Invalid Expression\n");
exit(0);
}
```

```python
# Compile the generated C files: lex.yy.c and y.tab.c
!gcc lex.yy.c y.tab.c -o art_expr -lfl
```

### C Lexical Analyzer

```python
%%writefile analyzer.c
#include<stdio.h>
#include<string.h>
#include<stdlib.h> // For exit()
#include<ctype.h>

FILE *fp;
char delim[] = { ' ', '\t', '\n', ',', ';', '(', ')', '{', '}', '[', ']', '#', '<', '>', '\0' };
char oper[] = { '+', '-', '*', '/', '%', '=', '!', '\0' };
char key[21][12] = {
    "int", "float", "char", "double", "bool", "void", "extern", "unsigned", "goto",
    "static", "class", "struct", "for", "if", "else", "return", "register", "long", "while", "do"
};
char predirect[2][12] = { "include", "define" };
char header[6][15] = { "stdio.h", "malloc.h", "process.h", "string.h", "ctype.h" }; // conio.h removed

void check(char[]);
int isdelim(char);
int isop_char(char); // Checks if a char is a *potential* start of an operator
void skipcomment(); // Stub for original function, no longer used directly

int main() {
    char fname[12];
    printf("\nEnter filename : ");
    scanf("%s", fname);
    fp = fopen(fname, "r");
    if (fp == NULL) {
        printf("\nThe file doesn’t exist.\n");
    } else {
        int c;
        while ((c = fgetc(fp)) != EOF) {
            if (isspace(c)) {
                continue;
            }
            // Handle comments
            if (c == '/') {
                int next_c = fgetc(fp);
                if (next_c == '/') { // Single line comment //
                    while ((c = fgetc(fp)) != EOF && c != '\n');
                    continue;
                } else if (next_c == '*') { // Multi-line comment /* ... */
                    int prev_c = 0;
                    while ((c = fgetc(fp)) != EOF && !(prev_c == '*' && c == '/')) {
                        prev_c = c;
                    }
                    continue;
                } else {
                    ungetc(next_c, fp); // Not a comment, treat '/' as operator
                    c = '/'; // Keep '/' for processing below
                }
            }
            // Handle string literals
            if (c == '"') {
                while ((c = fgetc(fp)) != EOF && c != '"');
                continue;
            }

            char token[50];
            int i = 0;

            if (isalpha(c) || c == '_') { // Start of an identifier
                token[i++] = c;
                while ((c = fgetc(fp)) != EOF && (isalnum(c) || c == '_')) {
                    if (i < 49) token[i++] = c; // Prevent buffer overflow
                    else break; // Token too long
                }
                ungetc(c, fp); // Put back the non-identifier char
                token[i] = '\0';
                check(token);
            } else if (isdigit(c)) { // Start of a number
                token[i++] = c;
                int has_decimal = 0;
                while ((c = fgetc(fp)) != EOF && (isdigit(c) || (c == '.' && !has_decimal))) {
                    if (c == '.') has_decimal = 1;
                    if (i < 49) token[i++] = c;
                    else break; // Token too long
                }
                ungetc(c, fp); // Put back the non-number char
                token[i] = '\0';
                check(token);
            } else if (isdelim(c)) {
                printf("\nDelimitter\t %c", c);
            } else if (isop_char(c)) {
                int next_c = fgetc(fp);
                // Check for two-character operators (e.g., ==, !=)
                if ((c == '=' && next_c == '=') || (c == '!' && next_c == '=')) {
                    printf("\nOperator\t %c%c", c, next_c);
                } else {
                    ungetc(next_c, fp); // Not a two-char operator, put it back
                    printf("\nOperator\t %c", c);
                }
            } else {
                printf("\nUnknown\t\t %c", c);
            }
        }
    }
    printf("\nEnd of file\n");
    if (fp != NULL) {
        fclose(fp);
    }
    return 0;
}

int isdelim(char c) {
    for (int i = 0; delim[i] != '\0'; i++) {
        if (c == delim[i])
            return 1;
    }
    return 0;
}

int isop_char(char c) {
    for (int i = 0; oper[i] != '\0'; i++) {
        if (c == oper[i])
            return 1;
    }
    return 0;
}

void check(char t[]) {
    if (strlen(t) == 0) return; // Should not happen with correct parsing logic

    // Check if it's a number (integer or float)
    int is_number = 1;
    int has_dot = 0;
    if (!isdigit(t[0])) {
        is_number = 0;
    } else {
        for (int i = 1; t[i] != '\0'; i++) {
            if (!isdigit(t[i])) {
                if (t[i] == '.' && !has_dot) {
                    has_dot = 1;
                } else {
                    is_number = 0;
                    break;
                }
            }
        }
    }

    if (is_number) {
        printf("\nNumber\t\t %s", t);
        return;
    }

    // Check if it's a preprocessor directive
    for (int i = 0; i < 2; i++) {
        if (strcmp(t, predirect[i]) == 0) {
            printf("\nPreprocessor directive %s", t);
            return;
        }
    }

    // Check if it's a header file
    for (int i = 0; i < 5; i++) { // Corrected array size
        if (strcmp(t, header[i]) == 0) {
            printf("\nHeader file\t %s", t);
            return;
        }
    }

    // Check if it's a keyword
    for (int i = 0; i < 21; i++) {
        if (strcmp(key[i], t) == 0) {
            printf("\nKeyword\t\t %s", key[i]);
            return;
        }
    }

    // Otherwise, it's an identifier
    printf("\nIdentifier\t %s", t);
}

void skipcomment() {
    // This function is no longer called directly. Its logic is integrated into main().
}
```
